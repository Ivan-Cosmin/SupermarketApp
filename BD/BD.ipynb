# [dbo].[AddProduct]

```sql
/****** Object:  StoredProcedure [dbo].[AddProduct]    Script Date: 23.05.2024 13:15:00 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[AddProduct]
    @Barcode VARCHAR(100),
    @Name VARCHAR(100),
    @CategoryName VARCHAR(50),
    @ManufacturerName VARCHAR(100),
    @ManufacturerCountryOfOrigin VARCHAR(100)
AS
BEGIN
    DECLARE @CategoryID INT;
    DECLARE @ManufacturerID INT;

    SELECT @CategoryID = ID
    FROM Categories
    WHERE Name = @CategoryName AND Deleted = 0;

    SELECT @ManufacturerID = ID
    FROM Manufacturers
    WHERE Name = @ManufacturerName AND CountryOfOrigin = @ManufacturerCountryOfOrigin AND Deleted = 0;


    INSERT INTO Products (Barcode, [Name], CategoryID, ManufacturerID)
    VALUES (@Barcode, @Name, @CategoryID, @ManufacturerID);
END;
GO
```

# [dbo].[GetAllProducts]

```sql
/****** Object:  StoredProcedure [dbo].[GetAllProducts]    Script Date: 23.05.2024 13:15:00 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[GetAllProducts]
AS
BEGIN
    SELECT 
        p.Barcode,
        p.Name,
        c.Name AS CategoryName,
        m.Name AS ManufacturerName,
        m.CountryOfOrigin AS ManufacturerCountryOfOrigin,
        p.Deleted
    FROM 
        Products p
    INNER JOIN 
        Categories c ON p.CategoryID = c.ID
    INNER JOIN 
        Manufacturers m ON p.ManufacturerID = m.ID
    WHERE c.Deleted = 0
        AND m.Deleted = 0;
END;
GO
```

# [dbo].[GetAllProductStocks]

```sql
/****** Object:  StoredProcedure [dbo].[GetAllProductStocks]    Script Date: 23.05.2024 13:15:00 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[GetAllProductStocks]
AS
BEGIN
    SET NOCOUNT ON;

    SELECT ps.ID, p.Name AS ProductName, p.Barcode, c.Name AS CategoryName, m.Name AS ManufacturerName,
           ps.Quantity, ps.UnitOfMeasure, ps.SupplyDate, ps.ExpiryDate, ps.PurchasePrice, ps.SellingPrice
    FROM ProductStocks ps
    INNER JOIN Products p ON ps.Barcode = p.Barcode
    INNER JOIN Categories c ON p.CategoryID = c.ID
    INNER JOIN Manufacturers m ON p.ManufacturerID = m.ID
    WHERE p.Deleted = 0 AND c.Deleted = 0 AND m.Deleted = 0;
END
GO
```

# [dbo].[GetReceiptID]

```sql
/****** Object:  StoredProcedure [dbo].[GetReceiptID]    Script Date: 23.05.2024 13:15:00 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[GetReceiptID]
    @CashierName VARCHAR(100),
    @Type VARCHAR(50),
    @Date DATE,
    @TotalPrice DECIMAL(10, 2)
AS
BEGIN
    DECLARE @CashierID INT;

    SELECT @CashierID = ID
    FROM Users
    WHERE Username = @CashierName
      AND Type = @Type
      AND Deleted = 0;

    SELECT ID
    FROM Receipts
    WHERE CashierID = @CashierID
      AND ReleaseDate = @Date
      AND TotalAmount = @TotalPrice;
END;
GO
```

# [dbo].[Login]

```sql
/****** Object:  StoredProcedure [dbo].[Login]    Script Date: 23.05.2024 13:15:00 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[Login]
    @Username NVARCHAR(50),
    @Password NVARCHAR(50),
    @Type NVARCHAR(50)
AS
BEGIN
    SET NOCOUNT ON;

    SELECT ID
    FROM Users
    WHERE Username = @Username
      AND Password = @Password
      AND Type = @Type
	  AND Deleted = 0;
END
GO
```

# [dbo].[ModifyProduct]

```sql
/****** Object:  StoredProcedure [dbo].[ModifyProduct]    Script Date: 23.05.2024 13:15:00 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[ModifyProduct]
    @Barcode VARCHAR(100),
    @NewBarcode VARCHAR(100),
    @NewName VARCHAR(100),
    @NewCategoryName VARCHAR(50),
    @NewManufacturerName VARCHAR(100),
    @NewManufacturerCountryOfOrigin VARCHAR(100)
AS
BEGIN
    -- Declarații pentru noile ID-uri
    DECLARE @CategoryID INT;
    DECLARE @ManufacturerID INT;

    -- Obținem CategoryID bazat pe noul nume de categorie
    SELECT @CategoryID = ID
    FROM Categories
    WHERE [Name] = @NewCategoryName AND Deleted = 0;

    -- Obținem ManufacturerID bazat pe noul nume de producător și țara de origine
    SELECT @ManufacturerID = ID
    FROM Manufacturers
    WHERE [Name] = @NewManufacturerName AND CountryOfOrigin = @NewManufacturerCountryOfOrigin AND Deleted = 0;
 
UPDATE Products
    SET 
        Barcode = @NewBarcode,
        Name = @NewName,
        CategoryID = @CategoryID,
        ManufacturerID = @ManufacturerID
    WHERE
        Barcode = @Barcode
END;
GO
```

# [dbo].[SearchProducts]

```sql
/****** Object:  StoredProcedure [dbo].[SearchProducts]    Script Date: 23.05.2024 13:15:00 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[SearchProducts]
    @Name VARCHAR(100) = NULL,
    @Barcode VARCHAR(100) = NULL,
    @ExpiryDate DATE = NULL,
    @ManufacturerName VARCHAR(100) = NULL,
    @CategoryName VARCHAR(50) = NULL
AS
BEGIN
    SELECT 
        p.Barcode,
        p.Name,
        s.ExpiryDate,
		s.Quantity AS Quantity,
		s.UnitOfMeasure AS UnitOfMeasure,
		s.SellingPrice AS SellingPrice,
        c.Name AS CategoryName,
        m.Name AS ManufacturerName,
        m.CountryOfOrigin AS ManufacturerCountryOfOrigin,
        p.Deleted
    FROM ProductStocks s
	JOIN
        Products p on p.Barcode = s.Barcode
    JOIN 
        Categories c ON p.CategoryID = c.ID
    JOIN 
        Manufacturers m ON p.ManufacturerID = m.ID
    WHERE 
        (@Name IS NULL OR p.Name LIKE '%' + @Name + '%') AND
        (@Barcode IS NULL OR p.Barcode LIKE '%' + @Barcode + '%') AND
        (@ExpiryDate IS NULL OR s.ExpiryDate = @ExpiryDate) AND
        (@ManufacturerName IS NULL OR m.Name LIKE '%' + @ManufacturerName + '%') AND
        (@CategoryName IS NULL OR c.Name LIKE '%' + @CategoryName + '%') AND
        p.Deleted = 0 AND
		c.Deleted = 0 AND
		s.Deleted = 0 AND
		m.Deleted = 0;
END;
GO
```

# [dbo].[UpdateStockQuantity]

```sql
/****** Object:  StoredProcedure [dbo].[UpdateStockQuantity]    Script Date: 23.05.2024 13:15:00 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[UpdateStockQuantity]
    @StockID INT,
    @Quantity INT
AS
BEGIN
    -- Declarați o variabilă pentru a stoca cantitatea curentă
    DECLARE @CurrentQuantity INT;

    -- Obțineți cantitatea curentă din stoc
    SELECT @CurrentQuantity = Quantity
    FROM ProductStocks
    WHERE ID = @StockID;


    -- Actualizați cantitatea din stoc
    UPDATE ProductStocks
    SET Quantity = Quantity - @Quantity
    WHERE ID = @StockID;
END;
GO
```
